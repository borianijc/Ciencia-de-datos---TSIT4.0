#¿Qué es Pandas profiling?
Pandas Profiling es una herramienta de análisis exploratorio de datos (EDA) en Python que ofrece una forma rápida y sencilla de obtener información detallada sobre un conjunto de datos.

Funciona en tres pasos principales:

 * Importación de datos: Puedes importar datos desde diversas fuentes, como archivos CSV, JSON, Excel o Pandas DataFrames.
 * Generación de informes: Analiza automáticamente los datos y genera un informe completo que incluye estadísticas descriptivas, distribuciones, correlaciones y más para cada variable en el conjunto de datos.
 * Visualización de resultados: Ofrece visualizaciones interactivas que ayudan a comprender mejor la estructura y las relaciones dentro de los datos, incluyendo histogramas, diagramas de dispersión, matrices de correlación y más.

#Características claves de Pandas Profiling:

* Facilidad de uso: No se necesita experiencia en programación para utilizarlo, ya que automatiza gran parte del proceso de EDA.
* Amplia gama de información: Proporciona una variedad de estadísticas descriptivas y visualizaciones para cada variable, lo que permite una comprensión exhaustiva de los datos.
* Personalización: Aunque es automatizado, permite cierto grado de personalización para adaptarse a las necesidades específicas del usuario.
* Comparación de conjuntos de datos: Permite comparar dos conjuntos de datos para identificar diferencias significativas en términos de distribuciones, correlaciones y más.


En resumen, Pandas Profiling es una herramienta poderosa para realizar análisis exploratorio de datos en Python, proporcionando información detallada y visualizaciones interactivas para comprender rápidamente la estructura y las relaciones dentro de los datos.

#¿Qué es Sweetviz?

Es una librería de Python que permite crear visualizaciones de datos de forma rápida y sencilla.
Es una herramienta muy útil para realizar análisis exploratorio de datos EDA.

Funciona mediante un proceso de tres pasos:


*   Importar los datos: se puede importar datos de una variedad de fuentes, incluyendo archivos CSV, JSON, Excel y Pandas DataFrames.

*   Seleccionar las características a visualizar: permite seleccionar las características que se desean visualizar.

*   Generar las visualizaciones: genera una variedad de visualizaciones, incluyendo diagramas de dispersión, histogramas, gráficos de barras y mapas de calor.

#Características de Sweetviz
Tiene una serie de características que la hacen una herramienta muy útil para la visualización de datos:

* Facilidad de uso: No se requiere experiencia en programación para crear visualizaciones de datos.

* Amplia gama de visualizaciones: Permite generar una variedad de visualizaciones.

* Personalización: Permite personalizar las visualizaciones para adaptarlas a las necesidades específicas del usuario final.
Comparación de conjuntos de datos: permite comparar dos conjuntos de datos (por ejemplo, conjuntos de entrenamiento y prueba) para identificar diferencias significativas en términos de distribuciones, estadísticas descriptivas y más.

```python
import pandas as pd
import sweetviz as sv
```

#Leemos el archivo para analizar

* En este caso, una base de datos de clientes de un servicio de telecomunicaciones, donde podemos ver el estado aquellos que permancen como clientes y los que se fueron, de estos, podemos ver los motivos y/o la razón por la cual solicitaron la baja del servicio.
* Vemos que el DataFrame contiene 7043 registros o filas por 8 columnas de datos.


**Se muestra el Data Frame con la cantidad de filas y columnas.**

```python
analisis_servicios2 = pd.read_csv("Telco_customer_churn_status.csv")
analisis_servicios2
```

**Con el comando .info() de pandas podemos ver las variables, la cantidad de datos no nulos que tienen y el tipo de datos que contienen.**
* Podemos ver como las variables 'Churn Category' y 'Churn Reason' tiene solo 1869 datos correctos, que son los registros de los clientes que dieron de baja el servicio. Los cuales a continuación haremos una extraccion de esos datos para realizar un analisis de los motivos y/o razones.

```python
analisis_servicios2.info()
```

**Con el comando .isnull anidando la función sum() podremos saber la cantidad de datos nulos que tiene la BBDD por variable.**
* Siguiendo lo mencionado anteiormente, comprobamos que las variables 'Churn Category' y 'Churn Reason' tiene 5174 datos nulos, que son los registros de los clientes que no dieron de baja el servicio.

```python
analisis_servicios2.isnull().sum().to_frame()
```

**A continuación con la función .dropna() y en este caso, seleccionando las variables de interés como 'Churn Category' y 'Churn Reason', eliminaremos los registros que sean nulos.**
* Luego de ejecutar podremos ver como ahora el DataFrame nos queda de 1869 filas o registros x las mismas 8 columnas que teníamos.

```python
analisis_servicios2 = analisis_servicios2.dropna(subset=['Churn Category', 'Churn Reason'])
analisis_servicios2
```

**Teniendo en cuanta las variables, como podremos ver en el informe propuesto por la librería pandas profiling, identificamos variables de las cuales podemos decir:**
* **Customer ID:** Es unique, ya que para cada cliente va a haber un único Id.
* **Count:** Como una constante en este caso, pero podría referirse a la cantidad de cuentas por usuario.
* **Quarter:** Es una variable categórica, se refiere al trimestre, en este caso todo el set es del Q3, pero si ampliamos la BBDD, podremos distinguir por trimestre del año.
* **Customer Status:** Variable categórica, que nos permite ver el estado del cliente, luego de la limpieza nos quedamos solo con aquellos que desconectaron el servicio.
* **Churn Label:** Es un dato Booleano donde indica si contiene datos del registro que dio de baja el servicio en formato (yes/no)  
* **Churn Value:** Mismo caso que el anterior, indica 0 si no esta desconectado y 1 si es de tipo Churned o desconectado.
* **Churn Category:** Variable categórica que nos indica la categoría de desconexión
* **Churn Reason:** Variable categórica donde podremos conocer las razones de desconexión.

**Utilizando Pandas, con la sigueinte sentencia vemos que la cantidad de registros corresponden a los clientes que dieron de baja el servicio.**

```python
customer_status = analisis_servicios2['Customer Status'].value_counts().to_frame()
customer_status
```

**Además, con Pandas podemos ver las categorías de las desconecciones y las razones de desconexion y la cantidad de registros por las distintas categorías y razones.**

```python
churned_category = analisis_servicios2['Churn Category'].value_counts().to_frame()
churned_category
```

```python
churned_reason = analisis_servicios2['Churn Reason'].value_counts().to_frame()
churned_reason
```


**Utilizando la Pandas Profiling, con el comando ProfileReport () nos muestra un reporte con todas las variables del DF para poder ver las posibilidades de analisis.**

```python
ProfileReport (analisis_servicios2)
```

**Ahora, utilizando la librería SweetViz, con el comando Analize () nos muestra un reporte con todas las variables del DF para poder ver las posibilidades de analisis.**

**Con el metodo show_notebook lo podemos ver en pantalla.**

```python
reporte = sv.analyze(analisis_servicios2)
reporte.show_notebook(h="full")
```
